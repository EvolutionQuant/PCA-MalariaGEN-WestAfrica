From https://vcftools.github.io/man_latest.html#INDIVIDUAL%20FILTERING%20OPTIONS

> --keep <filename>
>    
> --remove <filename>
> 
> Provide files containing a list of individuals to either include or exclude in subsequent analysis. Each individual ID (as defined in the VCF headerline) should be included on a separate line. If both options are used, then the "--keep" option is executed before the "--remove" option. When multiple files are provided, the union of individuals from all keep files subtracted by the union of individuals from all remove files are kept. No header line is expected.

````python
# save only index values
df_single_country_indices = pd.DataFrame(df_single_country.index)
df_single_country_indices
````

Download [ag1000g.phase2.ar1.pass.3L.vcf.gz](ftp://ngs.sanger.ac.uk/production/ag1000g/phase2/AR1/variation/main/vcf/pass/ag1000g.phase2.ar1.pass.3L.vcf.gz) (chromosome 3L) from [ftp://ngs.sanger.ac.uk/production/ag1000g/phase2/AR1/variation/main/vcf/pass/](ftp://ngs.sanger.ac.uk/production/ag1000g/phase2/AR1/variation/main/vcf/pass/)

## Position filtering


From https://vcftools.github.io/man_latest.html#INDIVIDUAL%20FILTERING%20OPTIONS

> --from-bp <integer>
> --to-bp <integer>
>
> These options specify a lower bound and upper bound for a range of sites to be processed. Sites with positions less than or greater than these values will be excluded. These options can only be used in conjunction with a single usage of --chr. Using one of these does not require use of the other.

## Compare new and original `vcf` files

Display the head of the original `vcf` file:

````python
%%bash
zcat ag1000g.phase2.ar1.pass.3L.vcf.gz | vcf-to-tab | head -n 5 >orig.tsv
````

````python
pd.read_csv("orig.tsv", sep='\t')
````

````python
%%bash
rm orig.tsv
````

Display the head of the new (recoded) `vcf` file:

````python
%%bash
cat ag1000g.phase2.ar1.pass.3L.Ghana.recode.vcf | vcf-to-tab | head -n 5 >recoded.tsv
````

````python
pd.read_csv("recoded.tsv", sep='\t')
````

````python
%%bash
rm recoded.tsv
````

## Validate the recoded `vcf` file

Compress `vcf` file to `vcf.gz`:

````python
%%bash
gzip -k ag1000g.phase2.ar1.pass.3L.Ghana.recode.vcf
ls -lh ag1000g.phase2.ar1.pass.3L.Ghana.recode.vcf*
````

Validate the recoded file:

````python
%%bash
vcf-validator -d ag1000g.phase2.ar1.pass.3L.Ghana.recode.vcf.gz
````
