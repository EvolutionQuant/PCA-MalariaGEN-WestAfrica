# Get IDs of mosquitoes from West Africa vs outgroup

````python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 199)
````

````python
df = pd.read_csv("samples.meta.txt", sep='\t', index_col=index_col)
df
````

Select only single mosquito species (*Anopheles gambiae*):

````python
populations = list(df["population"].unique())
populations
````

````python
if mosq_species:
    selection_mask = df["population"].str.contains(mosq_species)
    df = df[selection_mask]
df
````

````python
selected_populations = list(df["population"].unique())
selected_populations    
````

````python
countries = list(df["country"].unique())
countries
````

From https://en.wikipedia.org/wiki/West_Africa \[accessed on May 1, 2020\]

> The United Nations defines Western Africa as the 16 countries of Benin, Burkina Faso, Cape Verde, The Gambia, Ghana, Guinea, Guinea-Bissau, Ivory Coast, Liberia, Mali, Mauritania, Niger, Nigeria, Senegal, Sierra Leone and Togo, as well as the United Kingdom Overseas Territory of Saint Helena, Ascension and Tristan da Cunha

````python
west_afr = ["Benin", "Burkina Faso", "Cape Verde", "Gambia, The", "Ghana", "Guinea", "Guinea-Bissau", "Cote d'Ivoire", "Liberia", "Mali", "Mauritania", "Niger", "Nigeria", "Senegal", "Sierra Leone", "Togo", "the United Kingdom Overseas Territory of Saint Helena, Ascension and Tristan da Cunha"]
west_countries = [country for country in countries if country in west_afr]
outgroup = [country for country in countries if country not in west_afr]

# print(type(west_afr))
# print(type(countries))
# print(type(west_countries))
# print(type(outgroup))

print(f"west_afr == {west_afr}\n")
print(f"countries == {countries}\n")
print(f"west_countries == {west_countries}\n")
print(f"outgroup == {outgroup}")
````

Histogram of West African countries and the outgroup:

````python
histogram = {}
def update_histogram(key):
    if key not in histogram.keys():
        histogram[key] = 1
    else:
        histogram[key] += 1

for _, row in df.iterrows():

    if row["country"] in west_countries:
        update_histogram(row["country"])
    if row["country"] in outgroup:        
        update_histogram("outgroup")
    
print(f"histogram == {histogram}")
````

````python
# check total counts
total_individuals = len(df.index)
print(f"total_individuals == {total_individuals}")

hist_sum = 0
for val in histogram.values():
    hist_sum += val
print(f"hist_sum == {hist_sum} ")
````

````python
np_mask = np.array([False] * total_individuals)
# print(f"np_mask == {np_mask}, np_mask.shape == {np_mask.shape}\n")

histogram = {}       
for i, (_, row) in enumerate(df.iterrows()):
    hist_key = None
    if row["country"] in west_countries:
        hist_key = row["country"]
    elif row["country"] in outgroup:        
        hist_key = "outgroup"
        
    if hist_key:
        update_histogram(hist_key)
        if histogram[hist_key] <= max_indiv_per_group:
            np_mask[i] = True
    
print(f"histogram == {histogram}")

n_selected = np.sum(np_mask)
assert n_selected == len(histogram.keys()) * max_indiv_per_group
print(f"n_selected ({n_selected}) == "
      f"{len(histogram.keys())} * {max_indiv_per_group} == "
      f"({len(histogram.keys()) * max_indiv_per_group})")
````

````python
# select mosquitoes
df_west_and_outgroup = df[np_mask]
assert len(df_west_and_outgroup.index) == n_selected
df_west_and_outgroup
````

````python
# save only index values
df_indices = pd.DataFrame(df_west_and_outgroup.index)
df_indices
````

````python
# save selected Pandas dataframe to a file
outfilename = f"samples_west_and_outgroup.meta.{index_col}.txt"

print(f"Saving `df_indices` to {outfilename}")
df_indices.to_csv(outfilename, sep='\t', index=False, header=False)
````

````python
# display what is inside the saved file
df_outfile = pd.read_csv(outfilename, sep='\t', header=None)
df_outfile
````

## Individual filtering

From https://vcftools.github.io/man_latest.html#INDIVIDUAL%20FILTERING%20OPTIONS

> --keep <filename>
>    
> --remove <filename>
> 
> Provide files containing a list of individuals to either include or exclude in subsequent analysis. Each individual ID (as defined in the VCF headerline) should be included on a separate line. If both options are used, then the "--keep" option is executed before the "--remove" option. When multiple files are provided, the union of individuals from all keep files subtracted by the union of individuals from all remove files are kept. No header line is expected.

## Position filtering


From https://vcftools.github.io/man_latest.html#INDIVIDUAL%20FILTERING%20OPTIONS

> --from-bp <integer>
> --to-bp <integer>
>
> These options specify a lower bound and upper bound for a range of sites to be processed. Sites with positions less than or greater than these values will be excluded. These options can only be used in conjunction with a single usage of --chr. Using one of these does not require use of the other.
